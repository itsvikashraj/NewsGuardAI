```python


def get_onion_article_links():
    url = "https://www.newsbreak.com/"
    try:
        response = requests.get(url)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, "html.parser")

        article_links = set()
        for a in soup.find_all('a', href=True):
            href = a['href']

            if (
                href.startswith("https://www.newsbreak.com/")
            ):
                article_links.add(href.split("?")[0])

        return list(article_links)

    except Exception as e:
        print(f"Error: {e}")
        return []


if __name__ == "__main__":
    links = get_onion_article_links()
    print(f"🔗 Found {len(links)} article links:\n")
    for i, link in enumerate(links[:20], 1):
        print(f"{i}. {link}")
```
